```python
from sklearn import model_selection, preprocessing
import xgboost as xgb
xgb_params = {
    'eta': 0.05,
    'max_depth': 8,
    'subsample': 0.7,
    'colsample_bytree': 0.7,
    'objective': 'reg:linear',
    'eval_metric': 'rmse',
    'silent': 1
}
#training the model to get feature importance
dtrain = xgb.DMatrix(X_train, y_train, feature_names=X_train.columns.values)
model = xgb.train(dict(xgb_params, silent=0), dtrain, num_boost_round=100)
featureImportance = model.get_fscore()
```

```python
features = pd.DataFrame()
features['features'] = featureImportance.keys()
features['importance'] = featureImportance.values()
features.sort_values(by=['importance'],ascending=False,inplace=True)
fig,ax= plt.subplots()
fig.set_size_inches(20,10)
plt.xticks(rotation=90)
sns.barplot(data=features.head(15),x="importance",y="features",ax=ax,orient="h", color = "#34495e")
```

```python
classifier=Sequential()
classifier.add(Dense(12, input_dim=14,activation='relu',kernel_initializer='glorot_uniform'))

classifier.add(Dense(56,activation='relu',kernel_initializer='glorot_uniform'))
classifier.add(Dense(24,activation='relu',kernel_initializer='glorot_uniform'))
#classifier.add(Dense(24,activation='relu',kernel_initializer='glorot_uniform'))
classifier.add(Dense(1,activation='sigmoid',kernel_initializer='glorot_uniform'))
classifier.compile(optimizer='Nadam',loss='binary_crossentropy',metrics=['accuracy'])

```
